```python
test_string = str(preview)
print(type(test_string))
```

```python
print(test_string)
```

```python
#some langchain imp to split text re
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
#add params here for chunk_size, chunk_overlap, length_function=len, separators=[some list]
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 100,
    length_function=len,
    #markdown headers #H1, H2, H3
    separators= ["#", "##", "###"]
)
```

```python
chunks = text_splitter.split_text(test_string)
```

```python
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}:\n{chunk}\n")
```
